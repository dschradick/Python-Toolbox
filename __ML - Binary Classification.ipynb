### Coefficients, ORs, Conf-Ints

```python
def get_coefficients_table(model, alpha=.05, sort=True):
    params = model.params
    conf   = model.conf_int()

    conf['OR'] = params
    conf.columns = ['2.5%', '97.5%', 'OR']
    conf = np.exp(conf)

    conf['DIRECTION'] = ["+" if p >= 1 else "-" for p in conf.OR]
    conf['P_VALUE'] = model.pvalues
    conf.columns = ['2.5%', '97.5%','OR','DIRECTION','P_VALUE']
    conf = conf.round(3)

    significant = conf.query(f'P_VALUE < {alpha}')
    if sort: significant.sort_values('OR', ascending=False, inplace=True)
    return significant

significant = get_coefficients_table(model)
significant
```

```python
# FALSCH!!!!! Conf-ints auf OR-scale sind unsymmetrisch => tidy_logit benutzen!
stderr = np.abs(significant.iloc[:,2] - significant.iloc[:,1]) / 2 #??

plt.figure(figsize=(8,8))
plt.vlines(1,-1,20,linestyles='--',alpha=1,linewidth=1, color='steelblue', label='OR = 1 = No Effect')
plt.errorbar(significant['OR'], significant.index,  xerr=stderr, fmt='o', color='black',
             ecolor='black', capsize=5, capthick=True, elinewidth=.5);
plt.xlabel('Odds Ratio')
plt.legend();
```

```python
# from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, recall_score, precision_score, cohen_kappa_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import precision_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import auc
from sklearn.metrics import f1_score


def get_classification_result(y_true, predictions, thresholds):
    tns = []; fps = []; fns = []; tps = []
    precisions = []; accuracies = []; f1s = []
    
    for threshold in thresholds:
        label_pos = (predictions > threshold)
        tn, fp, fn, tp = confusion_matrix(y_test, label_pos).ravel()
        tns.append(tn); fps.append(fp); fns.append(fn); tps.append(tp)
        precisions.append(precision_score(y_test, label_pos))
        accuracies.append(accuracy_score(y_test, label_pos))
        f1s.append(f1_score(y_test,label_pos))

    result = pd.DataFrame({
        'threshold': thresholds, 
        'accuray': accuracies,
        'tn': tns,
        'fp': fps,
        'fn': fns,
        'tp': tps,
        'fpr':fpr,                    # TODO: explizit 
        'recall': tpr,                # explizit
        'precision': precisions,
        'f1': f1s}).round(3)

    return result.sort_values('threshold').query('threshold >=0 & threshold <= 1')

predictions = model.predict(X_test)

precision, recall ,thresholds = precision_recall_curve(y_test,predictions)
pr_auc = auc(recall, precision)

roc_auc = roc_auc_score(y_test, predictions)
fpr, tpr, thresholds =  roc_curve(y_test,predictions)

t_metrics = get_classification_result(y_test, predictions, thresholds)
```

### Confusion Matrix

```python
# Confusion Matrix
# => in numerischer Reihenfolge (0,1) & Predicted oben
#
#                    PREDICTED                  True...Rate                    False...Rate
#                     0    1                                                   = Error Rates 
#         -------------------------                                              (of Type I / II)
# ACTUAL  FALSE  |   TN    FP       Specificity = TNR = TN / (TN + FP)    ALPHA = FPR = FP / (TN + FP)  => Type I  Error Rate
#         TRUE   |   FN    TP       Sensitivity = TPR = TP / (TP + FN)    BETA  = FNR = FN / (TP + FN)  => Type II Error Rate                   
#
#                        Precision
#                      TP / (TP / FP)
#
# Power = Sensitiviy
#
# FPR = alpha = 1 - Specificity      => ROC is based on getting proba that prediced = 1 (either TP or FP) 
#                                       TP / P & FP / F
# FPR = 1 - TNR                      => uses the opposite => P and N 
#
# Type 1 Error = FP = predicting 1 altough truely 0
# Type 2 Error = FN = predicting 0 altough truely 1



from sklearn.metrics import plot_confusion_matrix

#plot_confusion_matrix(clf, X_test, y_test, cmap='Blues', values_format='.3g')
# => beschrifted auch gleich
# Für Threshold 0.8
confusion_matrix(y_test, (predictions > 0.8))
```
